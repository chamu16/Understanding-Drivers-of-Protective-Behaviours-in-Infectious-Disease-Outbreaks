# Data Cleaning

## YouGov Data

```python
# Extracting and renaming YouGov columns
yougov_1 = yougov[[
    "endtime",
    "qweek",
    "weight",
    "age",
    "gender",
    "state",
    "household_size",
    "employment_status",
    
    # Behaviour outcomes
    "i12_health_1",   # mask wearing
    "i12_health_2",   # washed hands
    "i12_health_3",   # used sanitiser
    "i12_health_6",   # avoided going out
    "i12_health_8",   # avoided public transport
    "i12_health_15",  # avoided crowded areas
    
    # Self-isolation
    "i9_health",
    "i11_health",
    
    # Behaviour intensity / mobility
    "i13_health",
    "i7a_health",
    
    # Perception / behavioural drivers
    "r1_1",
    "r1_2",
    "r1_6",
    "r1_7",
    
    # Trust
    "WCRex1"
]].rename(columns={
    "endtime": "date",
    "qweek": "survey_week",
    "weight": "survey_weight",
    "age": "age",
    "gender": "gender",
    "state": "state",
    "household_size": "household_size",
    "employment_status": "employment_status",

    "i12_health_1": "mask_wearing",
    "i12_health_2": "wash_hands",
    "i12_health_3": "use_sanitiser",
    "i12_health_6": "avoid_going_out",
    "i12_health_8": "avoid_public_transport",
    "i12_health_15": "avoid_crowded_areas",

    "i9_health": "would_self_isolate",
    "i11_health": "willingness_self_isolate",

    "i13_health": "handwashing_frequency",
    "i7a_health": "times_left_home",

    "r1_1": "covid_dangerous_for_me",
    "r1_2": "likely_to_get_covid",
    "r1_6": "important_to_improve_health",
    "r1_7": "life_affected_by_covid",

    "WCRex1": "trust_in_government_handling"
})
```

```python
# YouGov data info
yougov_1.info()
```

```python
# YouGov columns
yougov_1.columns
```

```python
# Convert date to datetime format
yougov_1["date"] = pd.to_datetime(
    yougov_1["date"],
    format="%d/%m/%Y %H:%M",
    errors="coerce"
).dt.normalize()
```

```python
# Select string/object columns
str_cols = yougov_1.select_dtypes(include=["object", "string"]).columns

# Remove leading and trailing whitespace
for col in str_cols:
    yougov_1[col] = yougov_1[col].astype("string").str.strip()
    
# Convert empty strings to missing values
yougov_1[str_cols] = yougov_1[str_cols].replace(r"^\s*$", pd.NA, regex=True)
```

```python
# Unique values in each columns
cols = [
    "gender",
    "state",
    "household_size",
    "employment_status",
    "mask_wearing",
    "wash_hands",
    "use_sanitiser",
    "avoid_going_out",
    "avoid_public_transport",
    "avoid_crowded_areas",
    "would_self_isolate",
    "willingness_self_isolate",
    "covid_dangerous_for_me",
    "likely_to_get_covid",
    "important_to_improve_health",
    "life_affected_by_covid",
    "trust_in_government_handling"
]

for col in cols:
    print(f"\nColumn: {col}")
    print(yougov_1[col].dropna().unique().tolist())
```

```python
# Replace common text missing values
missing_like = [
    "Don't know",
    "Prefer not to say",
    "Not sure",
]

yougov_1 = yougov_1.replace(missing_like, pd.NA)
```

```python
# Format column values
likert_cols = [
    "covid_dangerous_for_me",
    "likely_to_get_covid",
    "important_to_improve_health",
    "life_affected_by_covid"
]

for col in likert_cols:
    yougov_1[col] = (
        yougov_1[col]
        .astype("string")
        .str.extract(r"^\s*([1-7])", expand=False)
    )
```

```python
# Clean household_size
yougov_1["household_size"] = (
    yougov_1["household_size"]
    .astype("string")
    .str.replace("8 or more", "8", regex=False)
)
```

```python
# Define column groups
freq_cols = [
    "mask_wearing",
    "wash_hands",
    "use_sanitiser",
    "avoid_going_out",
    "avoid_public_transport",
    "avoid_crowded_areas",
]

yes_no_cols = [
    "would_self_isolate"
]

willing_cols = [
    "willingness_self_isolate"
]

trust_cols = [
    "trust_in_government_handling"
]

# Define mappings
freq_map = {
    "Not at all": 1,
    "Rarely": 2,
    "Sometimes": 3,
    "Frequently": 4,
    "Always": 5
}

yes_no_map = {
    "No": 0,
    "Yes": 1
}

willing_map = {
    "Very unwilling": 1,
    "Somewhat unwilling": 2,
    "Neither willing nor unwilling": 3,
    "Somewhat willing": 4,
    "Very willing": 5
}

trust_map = {
    "Very badly": 1,
    "Somewhat badly": 2,
    "Somewhat well": 3,
    "Very well": 4
}
```

```python
# Apply mappings
for col in freq_cols:
    yougov_1[col] = yougov_1[col].map(freq_map).astype("Int64")

for col in yes_no_cols:
    yougov_1[col] = yougov_1[col].map(yes_no_map).astype("Int64")

for col in willing_cols:
    yougov_1[col] = yougov_1[col].map(willing_map).astype("Int64")

for col in trust_cols:
    yougov_1[col] = yougov_1[col].map(trust_map).astype("Int64")
```

```python
# Convert to numeric data type
numeric_cols = [
    "household_size",
    "handwashing_frequency",
    "times_left_home",
    "covid_dangerous_for_me",
    "likely_to_get_covid",
    "important_to_improve_health",
    "life_affected_by_covid"
]

for col in numeric_cols:
    yougov_1[col] = yougov_1[col].astype("Int64")
```

```python
# Convert to category data type
cat_cols = [
    "gender",
    "state",
    "employment_status",
    "survey_week"
]

for col in cat_cols:
    yougov_1[col] = yougov_1[col].astype("category")
```

```python
# Value count
print("Handwashing frequecy:", (yougov_1["handwashing_frequency"] > 50).sum())
print("No. of times left home:", (yougov_1["times_left_home"] > 20).sum())
```

```python
# Replace unrealistic values with NaN
yougov_1.loc[yougov_1["handwashing_frequency"] > 50, "handwashing_frequency"] = np.nan
yougov_1.loc[yougov_1["times_left_home"] > 20, "times_left_home"] = np.nan
```

```python
# YouGov data info
yougov_1.info()
```

```python
# Check missing values (count + percentage)
missing_count = yougov_1.isna().sum()
missing_percent = round(((missing_count / len(yougov_1)) * 100), 2)

missing_summary_yougov = pd.DataFrame({
    "Missing_Count": missing_count,
    "Missing_Percentage": missing_percent
}).sort_values(by="Missing_Count", ascending=False)

print("Missing values by column (Count & Percentage):")
display(missing_summary_yougov)
```

```python
# Numeric columns
numeric_cols = [
    "household_size",
    "handwashing_frequency",
    "times_left_home"
]

# Ordinal columns
ordinal_cols = [
    "covid_dangerous_for_me",
    "likely_to_get_covid",
    "important_to_improve_health",
    "life_affected_by_covid",
    "trust_in_government_handling"
]

# Median imputation for numeric
for col in numeric_cols:
    yougov_1[col] = yougov_1[col].fillna(yougov_1[col].median())

# Mode imputation for ordinal
for col in ordinal_cols:
    mode_value = yougov_1[col].mode(dropna=True)[0]
    yougov_1[col] = yougov_1[col].fillna(mode_value)
```

```python
# Check missing values AFTER
print("Missing values AFTER:")
print(yougov_1.isna().sum().sort_values(ascending=False))
```

```python
# Sort first
oxcgrt_1 = oxcgrt_1.sort_values("date")
```

## OWID Data

```python
# Extracting and renaming OWID columns
owid_1 = owid[[
    "location",
    "date",
    "new_cases_smoothed",
    "new_cases_smoothed_per_million",
    "new_deaths_smoothed",
    "reproduction_rate",
    "people_fully_vaccinated_per_hundred",
    "hosp_patients",
    "icu_patients"
]].rename(columns={
    "location": "country",
    "date": "date",
    "new_cases_smoothed": "new_cases_smoothed",
    "new_cases_smoothed_per_million": "new_cases_smoothed_per_million",
    "new_deaths_smoothed": "new_deaths_smoothed",
    "reproduction_rate": "reproduction_rate",
    "people_fully_vaccinated_per_hundred": "people_fully_vaccinated_per_hundred",
    "hosp_patients": "hospital_patients",
    "icu_patients": "icu_patients"
})
```

```python
# Convert date and filter Australia
owid_1 = owid_1[owid_1["country"] == "Australia"].copy()
```

```python
# OWID data info
owid_1.info()
```

```python
# Convert date to datetime format
owid_1["date"] = pd.to_datetime(owid_1["date"])

# Drop extra column
owid_1 = owid_1.drop(columns=["country"])
```

```python
# OWID data info
owid_1.info()
```

```python
# Check missing values
missing_summary_owid = owid_1.isna().sum().sort_values(ascending=False)
print("Missing values by column:")
print(missing_summary_owid)
```

```python
# Sort by date
owid_1 = owid_1.sort_values("date").reset_index(drop=True)
```

```python
# Interpolation for time-series variables
interp_cols = [
    "new_cases_smoothed",
    "new_cases_smoothed_per_million",
    "new_deaths_smoothed",
    "reproduction_rate",
    "hospital_patients",
    "icu_patients"
]

# Interpolate and fill edge missing values
owid_1[interp_cols] = (
    owid_1[interp_cols]
    .interpolate(method="linear", limit_direction="both")
)

# VVaccination: 0 before rollout, then interpolate
owid_1["people_fully_vaccinated_per_hundred"] = (
    owid_1["people_fully_vaccinated_per_hundred"]
    .fillna(0)
    .interpolate(method="linear", limit_direction="both")
)
```

```python
# Check missing values AFTER
print("Missing values AFTER:")
print(owid_1.isna().sum().sort_values(ascending=False))
```
